```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Drop irrelevant columns

df = df.drop(["PassengerId", "Name", "Ticket", "Cabin"], axis = 1)

# These columns will not help much in result.
```

```python
# Handling Missing Values

df["Age"].fillna(df["Age"].median(), inplace = True)
df["Embarked"].fillna(df["Embarked"].mode()[0], inplace = True)
```

```python
# Convert Categorical columns into Numerical Columns

df = pd.get_dummies(df, columns=["Sex", "Embarked"], drop_first = True)
```

```python
# Define Features (X) & Target (y)

x = df.drop("Survived", axis = 1)
y = df["Survived"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Build & Train the model

from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(x_train, y_train)
```

```python
# Evaluate the model

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

y_pred = model.predict(x_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Define models
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(),
    "Support Vector Machine": SVC(),
    "K-Nearest Neighbors": KNeighborsClassifier()
}

# Train and test all
for name, model in models.items():
    model.fit(x_train, y_train)
    y_pred = model.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"{name} Accuracy: {acc:.4f}")
```

```python
# Plotting a confusion matrix

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
# Feature Importance (For Tree-based models)

import matplotlib.pyplot as plt

rf = RandomForestClassifier()
rf.fit(x_train, y_train)

# Plot
feat_importances = pd.Series(rf.feature_importances_, index=x.columns)
feat_importances.nlargest(10).plot(kind='barh')
plt.title("Top 10 Feature Importances (Random Forest)")
plt.show()
```

```python
# Using cross-validation to evaluate stability
# Using K-Fold cross-validation (usually cv=5) to check how stable your models are across splits.

from sklearn.model_selection import cross_val_score

score = cross_val_score(LogisticRegression(max_iter=1000), x, y, cv=5)
print("Logistic Regression CV Accuracy:", score.mean())
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "Random Forest": RandomForestClassifier(),
    "Support Vector Machine": SVC(),
    "K-Nearest Neighbors": KNeighborsClassifier()
}

for name, model in models.items():
    scores = cross_val_score(model, x, y, cv=5)
    print(f"{name} CV Accuracy: {scores.mean():.4f} (+/- {scores.std():.4f})")
```

```python
# Hyperparameter Tuning for Logistic Regression

from sklearn.model_selection import GridSearchCV

param_grid = {
    'C': [0.01, 0.1, 1, 10, 100],
    'solver': ['liblinear', 'lbfgs']
}

grid_lr = GridSearchCV(LogisticRegression(max_iter=1000), param_grid, cv=5)
grid_lr.fit(x, y)

print("Best Parameters (Logistic):", grid_lr.best_params_)
print("Best CV Score:", grid_lr.best_score_)
```

```python
# Hyperparameter Tuning for RandomForest - GridSearchCV

param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10]
}

grid_rf = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=5)
grid_rf.fit(x, y)

print("Best Parameters (RF):", grid_rf.best_params_)
print("Best CV Score:", grid_rf.best_score_)
```

```python
# Hyperparameter Tuning for SVC-GridSearchCV

param_grid = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

grid_svc = GridSearchCV(SVC(), param_grid, cv=5)
grid_svc.fit(x, y)

print("Best Parameters (SVC):", grid_svc.best_params_)
print("Best CV Score:", grid_svc.best_score_)
```

```python
# Hyperparameter Tuning for K-Nearest Neighbours-GridSearchCV

param_grid = {
    'n_neighbors': [3, 5, 7, 9],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan']
}

grid_knn = GridSearchCV(KNeighborsClassifier(), param_grid, cv=5)
grid_knn.fit(x, y)

print("Best Parameters (KNN):", grid_knn.best_params_)
print("Best CV Score:", grid_knn.best_score_)
```

```python
# Training Final Model with Best Params
# Using grid_rf.best_estimator_ to get the tuned model and retrain it.

best_rf = grid_rf.best_estimator_
best_rf.fit(x_train, y_train)
```

```python
# Evaluate on the Test Set

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

y_pred = best_rf.predict(x_test)

print("Test Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Feature importance plot

importances = pd.Series(best_rf.feature_importances_, index=x.columns)
importances.nlargest(10).plot(kind='barh')
plt.title("Top 10 Feature Importances")
plt.show()
```
